## Step 1: Import Necessary Libraries
complete the import statements below.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

## Step 2: Load the Dataset
Load the dataset (e.g., Iris dataset) or use your own.

```python
data = load_iris()
x = data.data
y = data.target
```

```python
data
```

## Step 3: Preprocess the Data (Standardization)
Standardize the dataset before applying PCA or LDA.

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

## Step 4: Apply PCA
Perform PCA and extract the principal components.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x_scaled)
```

```python
X_pca
```

## Step 5: Apply LDA
Perform LDA and extract the discriminant components.

```python
lda = LDA(n_components=2)
X_lda = lda.fit_transform(x_scaled, y)
X_lda
```

## Step 6: Visualization
Visualize PCA and LDA results using scatter plots.

```python
# Create a figure with 1 row and 2 columns of subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot PCA projection
ax1.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis')
ax1.set_title('PCA Projection')

# Plot LDA projection  
ax2.scatter(X_lda[:,0], X_lda[:,1], c=y, cmap='viridis')  
ax2.set_title('LDA Projection')

# Add some spacing between subplots
plt.tight_layout()
plt.show()
```

```python
# PCA is unsupervised and LDA is supervised.
# PCA will try to preserve the varience in columns whereas LDA will try to maximize the seperation between classes by target column.
# LDA will give better accuracy when we train our model
```

## Step 7: Feature Selection using Filter Methods
In this section, you will apply different filter methods such as:
- **Pearson Correlation Coefficient**
- **Chi-Square Test**
- **Variance Thresholding**

```python
import seaborn as sns
import numpy as np
correlation_matrix = np.corrcoef(x_scaled.T)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Feature Correlation Matrix')
plt.show()
```

### 8.1 Pearson Correlation Coefficient
Calculate the correlation between features and the target variable.

```python
from scipy.stats import pearsonr
import numpy as np

# Calculate correlations between features and target
correlations = []
for i in range(x.shape[1]):
    corr, _ = pearsonr(x[:, i], y)
    correlations.append((data.feature_names[i], abs(corr)))

# Sort features by correlation strength
correlations.sort(key=lambda x: x[1], reverse=True)
print("Pearson Correlations:")
for feature, corr in correlations:
    print(f"{feature}: {corr:.4f}")
```

### 8.2 Chi-Square Test
Use the Chi-Square test to determine the relationship between features and the target variable.

```python
from sklearn.feature_selection import SelectKBest, chi2

# Apply Chi-square test (need non-negative values)
x_normalized = StandardScaler().fit_transform(x)
x_normalized = x_normalized - x_normalized.min() # Make values non-negative

chi2_selector = SelectKBest(chi2, k=2) # Select top 2 features
x_chi2 = chi2_selector.fit_transform(x_normalized, y)

# Get selected feature indices
selected_features_chi2 = chi2_selector.get_support(indices=True)
print("\nChi-Square Selected Features:")
print([data.feature_names[i] for i in selected_features_chi2])
```

### 8.3 Variance Thresholding
Remove features with low variance as they may not be useful for predictions.

```python
from sklearn.feature_selection import VarianceThreshold

# Remove features with low variance
selector = VarianceThreshold(threshold=0.1)
x_var = selector.fit_transform(x)

# Get selected feature indices
selected_features_var = selector.get_support(indices=True)
print("\nVariance Thresholding Selected Features:")
print([data.feature_names[i] for i in selected_features_var])
```

```python
import matplotlib.pyplot as plt

methods = ['Original', 'Pearson (top 2)', 'Chi-Square (top 2)', 'Variance (>0.1)']
feature_counts = [x.shape[1], 2, 2, len(selected_features_var)]

plt.figure(figsize=(10, 5))
plt.bar(methods, feature_counts)
plt.title('Number of Features Selected by Different Methods')
plt.xticks(rotation=45)
plt.ylabel('Number of Features')
plt.tight_layout()
plt.show()
```

## Step 10: Applying PCA on Digits Dataset
In this section, we will apply PCA on the `load_digits` dataset, which consists of handwritten digit images, and visualize its effect.

### 10.1 Load and Visualize the Digits Dataset
We will load the dataset and visualize some sample images.

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()
X_digits = digits.data
y_digits = digits.target

fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f'Label: {digits.target[i]}')
    ax.axis('off')
plt.show()

# or
#plt.imshow(digits.images[0], cmap='gray')
```

```python
# Every time imshow will use to visualize the pixel images
```

### 10.2 Apply PCA and Reduce Dimensions
We will reduce the dataset to 2 principal components and visualize the transformed data.

```python
from sklearn.decomposition import PCA
```

```python
pca_digits = PCA(n_components=2)
X_pca_digits = pca_digits.fit_transform(X_digits)
```

```python
x_train_pca, x_test_pca, y_train_pca, y_test_pca = train_test_split(X_digits, y_digits, test_size=0.20, random_state=0)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
lda_digits = LDA(n_components=2)
X_lda_digits = lda_digits.fit_transform(X_digits, y_digits)
lda_digits
```

```python
x_train_lda, x_test_lda, y_train_lda, y_test_lda = train_test_split(X_lda_digits, y_digits, test_size=0.20, random_state=0)
```

```python
# Create a figure with 1 row and 2 columns of subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot PCA projection
ax1.scatter(X_pca_digits[:, 0], X_pca_digits[:, 1], c=y_digits, cmap='viridis')
ax1.set_title('PCA Projection of Digits')

# Plot LDA projection
ax2.scatter(X_lda_digits[:, 0], X_lda_digits[:, 1], c=y_digits, cmap='viridis')
ax2.set_title('LDA Projection of Digits')

# Add some spacing between subplots
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_digits = RandomForestClassifier()
X_rf_digits = rf_digits.fit(x_train_lda, y_train_lda)

X_rf_digits.score(x_test_lda, y_test_lda)
```

```python
# Get the image at index 1796
test_image = X_lda_digits[1796].reshape(1, -1)  # Reshape to match input format

# Make prediction
prediction = rf_digits.predict(test_image)
print(f"Predicted digit: {prediction[0]}")

# Optional: Display the actual label for comparison
actual = y_digits[1796]
print(f"Actual digit: {actual}")

# Optional: If you want to visualize the image
plt.imshow(X_digits[1796].reshape(8, 8), cmap='gray')
plt.title(f'Predicted: {prediction[0]}, Actual: {actual}')
plt.axis('off')
plt.show()
```

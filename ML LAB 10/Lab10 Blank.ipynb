# Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs, make_moons
from sklearn.preprocessing import StandardScaler
```

# Step 2: Generate dummy data using make_blobs

```python
x, y = make_blobs(n_samples = 300, centers=3, random_state=42)
x = StandardScaler().fit_transform(x)
```

# Step 3: Plot Generated data 

```python
plt.scatter(x[::,0], x[::,1])
plt.title("Raw Data")
plt.show()
```

# Step 4: Apply Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
model.fit(x)
```

```python
model.linkage #ward
model.labels_
```

# Step 5: Display Dendrogram

```python
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(x, method='ward')
plt.figure(figsize=(10,5))
dendrogram(linked)
plt.show()
```

# Step 6: Generate dummy data using make_moons

```python
x, y = make_moons(n_samples = 300, noise=0.09, random_state=42)
x = StandardScaler().fit_transform(x)
```

# Step 7: Plot Generated data

```python
plt.scatter(x[::,0], x[::,1])
plt.title("Raw Data")
plt.show()
```

# Step 8: apply DBSCAN Clustering 

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c=dbscan.labels_)
```

# Step 9: apply K Mean Clustering on same data

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(x)
centerdata = kmeans.cluster_centers_
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c=model.labels_)
plt.scatter(centerdata[:,0], centerdata[:,1], marker="X", c="red")
```

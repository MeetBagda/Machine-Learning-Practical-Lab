## Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

## Step 2.  Read the Data

```python
dfsalary = pd.read_csv("./Salary_Data.csv")
```

```python
dfsalary
```

## Step 3. View Data using Scatter plot 

```python
plt.scatter(dfsalary['YearsExperience'],dfsalary['Salary'])
```

## Step 4. Splitting dataset in to input and output

```python

A = dfsalary['YearsExperience'].values.reshape(-1,1)
B = dfsalary['Salary']
A
```

## Step 4. Splitting dataset in to input and output

```python
A_train,A_test,B_train,B_test = train_test_split(A,B,train_size=0.8)
```

## Step 6. Create LinearRegression Model

```python
model_salary = linear_model.LinearRegression()
```

## Step 7. Fit the Model (Training)

```python
model_salary.fit(A_train,B_train)
```

##  Step 6. Predicting the Test set results

```python
model_salary.predict([[60]])
NewB= model_salary.predict(A_test)
NewB
```

##  Step 7. Visualising the Training set results

```python
plt.scatter(A_train,B_train)
```

##  Step 8. Visualising the Test set results

```python
plt.scatter(A_test,B_test)
plt.plot(A_test,NewB,marker="*")
model.score(A_test,B_test)
```

# Faculty Salary Prediction using beautifulsoup4

```python
import requests
from bs4 import BeautifulSoup 
```

```python
r = requests.get('https://www.darshan.ac.in/faculty-list/btech-computer')

# success code - 200
print(r)
```

```python
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python

s = soup.find('div', class_='row justify-content-center justify-content-sm-start g-pb-25')

facname = s.find_all('h2', class_='h5 g-font-primary g-color-black g-font-weight-600 text-center g-mb-0')
# print(facname)

facexp = s.find_all('span', class_='btn btn-sm g-bg-blue-opacity-0_1 g-color-blue g-font-weight-700 g-font-size-12')
# print(facexp)


# Create a mapping of faculty names to their experience
faculty_info = {}
for name, exp in zip(facname, facexp):
    faculty_info[name.text.strip()] = exp.text.strip()

# Print the mapping
for name, exp in faculty_info.items():
    print(f"Faculty Name: {name}, Experience: {exp}")
```

```python
import csv
# Create a list of dictionaries for the data
faculty_data = []
for name, exp in zip(facname, facexp):
    faculty_data.append({'Faculty Name': name.text.strip(), 'Experience': exp.text.strip()})

# Save the data to a CSV file
csv_file_name = 'faculty_scraped_data.csv'

# Writing to CSV
with open(csv_file_name, mode='w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['Faculty Name', 'Experience']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    writer.writeheader()  # Write the header row
    writer.writerows(faculty_data)  # Write the data rows

print(f"Data successfully saved to {csv_file_name}")
```

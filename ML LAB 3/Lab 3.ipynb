## Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

## Step 2. Import the dataset

```python
df = pd.read_csv("./50_Startups.csv")
df
```

## Step 3 . Check the State Column

```python
df.State
```

```python
df.plot()
```

## Step 4 . Splitting dataset in to input and output

```python
X = df.iloc[::,0:4:]
Y = df.iloc[::,4:5:]
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Import  ColumnTransformer

```python
from sklearn.compose import ColumnTransformer # lengthy process
```

## Step 5.2 .  Import  OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder # lengthy process
```

## Step 5.3 .  Perform Transformation

```python
# X1 = pd.get_dummies(X, columns=['State']) , # return boolean values, it will also accept in fit method.
X1 = pd.get_dummies(X, columns=['State'],dtype='int') # return int values

X1
```

```python
# But after converting column, it will create :
# redundancy
# corelation/linearity
# model train unwell
```

## Step 6 . Dummy variable trap

```python
X1 = pd.get_dummies(X, columns=['State'],dtype='int',drop_first=True) # drop one column
X1
```

## Step 7 Splitting dataset in to Train and Test

```python
X_train,X_test,Y_train,Y_test = train_test_split(X1,Y,train_size=0.8,random_state=1)
```

## Step 8  Import LinearRegression model from linear_model family 

```python
from sklearn import linear_model

model_startup = linear_model.LinearRegression()
```

## Step 9  Fit the data

```python
model_startup.fit(X_train,Y_train)
```

## Step 10  Predict the data

```python
NewY = model_startup.predict(X_test)
NewY
```

## Step 11  Display Result

```python
plt.scatter(X_test['Administration'],Y_test)
plt.scatter(X_test['Administration'],NewY)
```
